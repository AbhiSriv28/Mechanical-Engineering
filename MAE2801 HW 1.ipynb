```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x=np.linspace(-100,100,num=200)
y=np.linspace(-100,100,num=200)
xx, yy = np.meshgrid(x,y)
```

```python
z = -(xx+2*yy)/3
z1=-(3*xx+2*yy)
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(xx, yy, z, alpha=0.5,label = "P1")
ax.plot_surface(xx, yy, z1, alpha=0.5, label = "P2")

x, y, z = [50,-50], [-100,100], [50,-50]
ax.scatter(x, y, z, c='black', s=1)
ax.plot(x, y, z, color='black')
xz=[0]
yz=[0]
zz=[0]

ax.scatter(xz,yz,zz, c ='red', s=10)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```
