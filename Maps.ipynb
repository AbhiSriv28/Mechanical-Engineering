```python
import folium
from datascience import *
import pandas as pd
import numpy as np
df = pd.read_csv('Site_list.csv')
df
#importing data in this block
```

```python
df.dropna(subset=['SNO'],inplace=True)
df
#cleaning up data to remove NaN from Lat and Long else Folium gives error
```

```python
my_map1 = folium.Map(location=[29.7604,-95.3698 ], zoom_start=10)
#baseline map with lat and long taken as center of Houston city from google maps
i=0
while i<df["SNO"].iloc[-1]:
 folium.Marker(location=[df['Latitude'][i], df['Longitude'][i]], popup=df['SITE'][i]).add_to(my_map1)
 i=i+1
my_map1
```
